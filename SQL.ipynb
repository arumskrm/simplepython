```python
#Q1 (Which country has the lowest death caused by nutrition in 2019?)
nutrisi = ['Cause of death, by communicable diseases and maternal, prenatal and nutrition conditions (% of total)']
df_nutrisi = df[df['Indicator Name'].isin(nutrisi)]
df_nutrisi.loc[:,['Country Name','2019']].sort_values('2019',ascending=False)
```

```python
#Q2 (Which country has the highest rate of diabetes prevalence in 2021?)
diabetes = ['Diabetes prevalence (% of population ages 20 to 79)']
df_diabetes = df[df['Indicator Name'].isin(diabetes)]
df_diabetes.loc[:,['Country Name','2021']].sort_values('2021',ascending=False)
```

```python
#Q3 (How much is the working age population of Mali in 2020?)
mali = ['Mali']
df_mali = df[df['Country Name'].isin(mali)]
working=['Age dependency ratio (% of working-age population)']
df_mali_working=df_mali[df_mali['Indicator Name'].isin(working)]
df_mali_working.loc[:,['Country Name','2020']].sort_values('2020',ascending=False)
```
